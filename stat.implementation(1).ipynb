```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#normal distribution / gaussion dist

mu, sigma = 0 , 0.1
s=np.random.normal(mu,sigma,size = 1000)
print("count of dps", len(s))
s
```

```python
plt.hist(s,100,density = True)
plt.show()
```

```python
sns.distplot(s)
```

```python
# uniform dist> all the outcomes has equal prob to occur in range

u = np.random.uniform(0,1,100)

plt.hist(u,10)
plt.show()
u
```

```python
# log normal dist>> whose logarithm is normally distributrd - Right skewed
l = np.random.lognormal(mean = 0 , sigma = 0.5, size = 1000)
l=l[np.where(abs(l) < 5)]
plt.hist(l,bins = 100)
plt.show()
```

```python
# Log normal --> normal --> log

sns.distplot(np.log(l))
```

```python
# Log normal --> normal --> square root

sns.distplot(np.sqrt(l))
```

```python
# chisquare distribution

k=np.random.chisquare(df =1 , size = 1000)
k = k[np.where(abs(k) <10 )]
plt.hist(k,bins = 100)
plt.show()
```

```python
# t distribution

t=np.random.standard_t(df =1, size = 1000)
t = t[np.where(abs(t) <10 )]
plt.hist(t,bins = 100)
plt.show()
```

```python
# t dist -- dof = 10 --tail is thinner/smaller
t=np.random.standard_t(df =100, size = 1000)
t = t[np.where(abs(t) <10 )]
plt.hist(t,bins = 100)
plt.show()
```

```python
#  f distribution
# dfum is df for numerator, dfden is df for den
f=np.random.f(dfnum =1,dfden= 1, size = 1000)
f = f[np.where(abs(f) <10 )]
plt.hist(f,bins = 100)
plt.show()
```
